#**CENTRAL LIMIT THEOREM**

The Central Limit Theorem (CLT) is the statistical theory that says that when a sufficiently large sample is taken from a population, the mean of all samples in the same population will approach the mean of the population. Accordingly, in a dataset that does not show normal distribution, the distribution can be normalized by increasing the sample size. 


As seen in the definition and formula below, all samples will follow an approximate normal distribution pattern, with all variances approximately equal to the variance of the population divided by the size of each sample. In samples containing “n” number of independent variables, the distribution approaches N(0,1) as “n” goes to infinity.*

#***How We Understand the Distribution NOT Normal?***

In order to understand whether the distribution is normal, a histogram of the data is drawn using the normality test. If the skewness is high in the histogram, it is far from normal. In the histogram, which gives a symmetrical bell curve relative to the mean of the distribution, the ɑ-value is 0.05 at the 95% confidence interval. When the P-value is less than the ɑ-value, the null hypothesis (H0), which states that the distribution is normal in the determined confidence interval, is rejected, since the sample data will fall outside the 95% acceptance area in the histogram.

Here in below, we are going to prove Central Limit Theorem by using Numpy with random data.

```python
# we create a population distribution

import numpy as np
import random

shape, scale = 3., 2. # mean and width
s = np.random.uniform(low = -10, high = 10, size = 1000) # randomly uniform data between -10 and 10

import matplotlib.pyplot as plt
import scipy.special as sps
plt.hist(s)
plt.show()

```

```python
# we take samples randomly from the population

plt.hist(random.choices(s, k=100))
plt.show()
```

```python
# ctl is just not the taking samples, it says that when the sample size gets bigger, mean of samples gets closer to the population mean.

samples = [np.mean(random.choices(s, k=10)) for _ in range(50)] # we take samples for fifty times
plt.hist(samples)
plt.show()
```

```python
# we take samples for hundred times

samples = [np.mean(random.choices(s, k=10)) for _ in range(100)] 
plt.hist(samples)
plt.show()
```

```python
# we take samples for five hundred times

samples = [np.mean(random.choices(s, k=10)) for _ in range(500)] 
plt.hist(samples)
plt.show()
```

As we have seen above with application, when we increase the sample size, distribution approaches the normal distribution.
